# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    visited = []

    for node in graph:
        distances[node] = float('inf')

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance

    return distances
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```

```python
import heapq

def bellman_ford(graph, V, src):
    dist = {v: float("inf") for v in graph}
    parent = {v: None for v in graph}
    dist[src] = 0

    # Relax edges V-1 times
    for _ in range(V - 1):
        for u in graph:
            for v, w in graph[u].items():
                if dist[u] != float("inf") and dist[u] + w < dist[v]:
                    dist[v] = dist[u] + w
                    parent[v] = u

    # Check for negative cycles
    for u in graph:
        for v, w in graph[u].items():
            if dist[u] != float("inf") and dist[u] + w < dist[v]:
                print("Graph contains a negative weight cycle!")
                return None, None

    return dist, parent


def dijkstra(graph, src):
    dist = {v: float("inf") for v in graph}
    parent = {v: None for v in graph}
    dist[src] = 0
    pq = [(0, src)]

    while pq:
        d, u = heapq.heappop(pq)
        if d > dist[u]:
            continue
        for v, w in graph[u].items():
            if dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                parent[v] = u
                heapq.heappush(pq, (dist[v], v))

    return dist, parent


def print_path(parent, dest):
    path = []
    while dest is not None:
        path.append(dest)
        dest = parent[dest]
    return path[::-1]


# --- MAIN PROGRAM ---
# Example graph input using dictionary
graph = {
    0: {1: 4, 2: 5},
    1: {2: -3, 3: 2},
    2: {3: 3},
    3: {4: 2},
    4: {5: -1},
    5: {}  # new node added
}

# Add two more nodes
graph[6] = {3: 1, 5: 4}
graph[7] = {6: 2}

V = len(graph)

# Test with different sources
for src in [0, 1, 6]:
    dest = 5
    print(f"\nSource: {src}, Destination: {dest}")

    print("--- Bellman-Ford ---")
    dist_bf, parent_bf = bellman_ford(graph, V, src)
    if dist_bf:
        print(f"Distance: {dist_bf[dest]}")
        print("Path:", print_path(parent_bf, dest))

    print("--- Dijkstra ---")
    dist_dij, parent_dij = dijkstra(graph, src)
    print(f"Distance: {dist_dij[dest]}")
    print("Path:", print_path(parent_dij, dest))
```
