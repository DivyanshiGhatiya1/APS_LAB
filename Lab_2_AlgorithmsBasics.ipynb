```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
import time
start=time.time()
print(func4([1, 2, 3, 4], [2, 4, 6, 8]))
end=time.time()
print((end-start)*1000)
#This algorithm counts how many elements of list a are also present in list b
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Sample function call: func5(16)
import time
start=time.time()
p=print(func5(16))
end=time.time()
print((end-start)*1000)
#this function produces square root of a no
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
#Sample function call: func8([3,1,2,7,8,6])
import time
start=time.time()
p=print(func8([3,1,2,7,8,6]))
end=time.time()
print((end-start)*1000)
#This algorithm recursively sorts the array using a divide-and-conquer approach
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
#Sample function call: func12([1, 2, 3, 4, 5], 2)
import time
start=time.time()
print(func12([1, 2, 3, 4, 5], 2))
end=time.time()
print((end-start)*1000)
#This function rotates the array to the right by k positions
```

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
  sum=0
  for i in arr:
    sum=sum+i
  return sum
```

```python
#call the function
arr = [12, 3, 4, 15,45,78,450]
import time
start=time.time()
print(sum_arr(arr))
end=time.time()
print((end-start)*1000)

```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(array,key):
  count=0
  for i in array:
    if(i==key):
      count=1
      break
  if count==1:
      print(key,"is present in array")
  else:
    print(key,"is not present in array")
```

```python
# call the function liner_Search for array = [1, 3, 5, 7, 9]
array = [1, 3, 5, 7, 9]
key=int(input("enter key:"))
import time
start=time.time()
linear_Search(array,key)
end=time.time()
print((end-start)*1000)
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(array,key):
    low =0
    high=len(arr)-1
    while low<=high:
        mid = (low+high)
        if arr[mid]==key:
            return mid
        elif arr[mid]<key:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# call the function Binary_Search for array arr = [1, 3, 5, 7, 9,15,16,14,45]
arr = [1, 3, 5, 7, 9,15,16,14,45]
key=int(input("enter target:"))
import time
start=time.time()
print(Binary_Search(array,key))
end=time.time()
print((end-start)*1000)
```
